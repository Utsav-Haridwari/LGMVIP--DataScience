```python
# Importing Image Processing libraries
from PIL import Image
from IPython.display import display
import cv2
```

```python
img = cv2.imread(r'C:\Users\utsaj haridwari\Documents\Internship\Lets Grow More\Tasks\Task 4\mountain.jpg', 1)
img = cv2.resize(img, (640, 400))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img))
# Reading the image and then resizing it.
```

```python
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(img_grey))
# Converting into greyscale
```

```python
img_inv = cv2.bitwise_not(img_grey)
display(Image.fromarray(img_inv))
# Inverted Image
```

```python
img_smooth = cv2.GaussianBlur(img_inv, (5, 5), 0, 0)
display(Image.fromarray(img_smooth))
# Smoothing it
```

# Inverting the image to get final sketch

```python
def Sketch(img_grey, img_smooth):
    img_smooth_inv = 255 - img_smooth
    return cv2.divide(img_grey, img_smooth_inv, scale=256.0)
# Creating a function for sketching
```

```python
img_final = Sketch(img_grey, img_smooth)
display(Image.fromarray(img_final))
# Sketched image
```

```python
display(Image.fromarray(img))
display(Image.fromarray(img_final))
# Comparing Image and Sketch
```
