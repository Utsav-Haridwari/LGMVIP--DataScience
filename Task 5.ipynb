# Choosing the required columns for analysing the data

```python
data=data[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
           "attacktype1_txt","targtype1_txt","success","gname","motive","nkill","nwound"]]
```

```python
data["nkill"]=data["nkill"].fillna(0)
data["nwound"]=data["nwound"].fillna(0)
data["No of affected people"]=data["nkill"]+data["nwound"]
```

```python
data.describe()
```

# Maximum no. of killed person are 1570

# Analysing the number of affected people per year

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,15))
plt.xlabel("Years",fontsize=30)
plt.ylabel("No of affected people",fontsize=30)
data.groupby('iyear')["No of affected people"].sum().plot.bar()
plt.show()
```

# In 2014, Maximum number of people were affected(i.e both killed and wounded) by terrorist attack 

# Analysing the number of attacks by country(Top 15 Countries)

```python
import numpy as np

Country_attack=data["country_txt"].value_counts().head(15) 
Country_attack
```

```python
Country_attack.plot(kind="bar", color='orange', figsize=(20,10))
plt.title("Analyzing No of attacks by country", fontsize=20)
plt.xlabel("Country", fontsize=20)
plt.ylabel("No of Attacks", fontsize=20)
plt.show()
```

# Analysing the number of attacks by Region

```python
plt.pie(data["region_txt"].value_counts(), radius=3, autopct="%1.1f%%", labels=data["region_txt"].value_counts().index)
plt.show()
```

# Analysing the number of attacks by city(Top 20 cities)

```python
city=data["city"].value_counts() [1:21]
```

```python
city.plot (kind="bar", color='black', figsize=(20,10))
plt.title("Analyzing No of attacks by city", fontsize=20)
plt.xlabel("city", fontsize=20) 
plt.ylabel("No of Attacks", fontsize=20)
plt.show()
```

# Analysing the number of attacks by States(Top 20 States)

```python
state=data["provstate"].value_counts().head(20)
state
```

```python
state.plot(kind="bar",color='pink', figsize=(20,10))
plt.title("Analyzing No of attacks by state", fontsize=20) 
plt.xlabel("states", fontsize=20) 
plt.ylabel("No of Attacks", fontsize=20) 
plt.show()
```
